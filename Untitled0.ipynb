```python
#Toribio Torres Marifer
#Codigo: 216850837
#SecciÃ³n D02
import threading
import time
import queue
import random

veces_comer = 0
 
class Filosofo(threading.Thread):
    def __init__(self, ID):
        super().__init__()
        self.ID = ID
        
    def run(self):
        print(f"\n\n[*] Filosofo {self.ID + 1} PENSANDO [*]")
        time.sleep(0.05)

        while True:
            if filosofo_comer[self.ID] == veces_comer: #Filosofo ya comio numero de veces ingresada por el usuario
                return 
            
            if comiendo.full(): #Ya estan comiendo dos filosofos, el filosofo hambriento aun no puede comer
                time.sleep(random.randint(5,7))
        
            else:
                cantidad_comiendo = len(filosofos_comiendo) #Si ya hay un filosofo comiendo
                if cantidad_comiendo > 0:
                    self.comprobarComer() #Verificar que puede comer, que no sea un filosofo que este al lado del que esta comiendo
                            
                else: #No hay filosofos comiendo
                    self.puedeComer() #El filosofo puede comer
            
    def comprobarComer(self): #Comprobar que el filosofo hambriento, no este al lado del filosofo que ya esta comiendo
        puedeComer = False
        if self.ID == 0:
            if filosofos_comiendo[0] != 4 and filosofos_comiendo[0] != 1:
                puedeComer = True
                self.puedeComer()
        
        elif self.ID == 1:
            if filosofos_comiendo[0] != 0 and filosofos_comiendo[0] != 2:
                puedeComer = True
                self.puedeComer()
            
        elif self.ID == 2:
            if filosofos_comiendo[0] != 1 and filosofos_comiendo[0] != 3:
                puedeComer = True
                self.puedeComer()
            
        elif self.ID == 3:
            if filosofos_comiendo[0] != 2 and filosofos_comiendo[0] != 4:
                puedeComer = True
                self.puedeComer()
    
        elif self.ID == 4:
            if filosofos_comiendo[0] != 0 and filosofos_comiendo[0] != 3:
                puedeComer = True
                self.puedeComer()
            
        if puedeComer == False:
            time.sleep(random.randint(5,7))
    
    def puedeComer(self): #El filosofo puede empezar a comer
        filosofos_comiendo.append(self.ID) #Agregar filosofo a lista de comiendo
        comiendo.put(self, block=False) #Agregar filosofo a cola de comiendo
        self.empezarComer() #LLamar metodo empezarComer, para que filosofo pueda comer

    def empezarComer(self):
        print(f"\n\n[+] Filosofo {self.ID + 1} HAMBRIENTO [+]")
        print(f"\n\n[~] El filosofo {self.ID + 1} ESTA COMIENDO [~]")
        time.sleep(0.5)
        filosofo_comer[self.ID] = filosofo_comer[self.ID] + 1 #Incrementar veces que comio filosofo
        print(f"\n\n[x] El filosofo {self.ID + 1} TERMINO DE COMER [x] *** VECES QUE HA COMIDO: {filosofo_comer[self.ID]} ***")
        
        if len(filosofos_comiendo) > 0: #Hay filosofos en la lista de filosofos comiendo
            comiendo.get() #Sacar filosofo de la cola
            filosofos_comiendo.pop(0) #Sacar filosofo de la lista

        self.pensar() 
        
    def pensar(self):
        print(f"\n\n[*] Filosofo {self.ID + 1} PENSANDO [*]")
        time.sleep(random.randint(5,7))
        
if __name__ == '__main__':
    print("|||||||||||||||||||||||||||||||||||||||||||||||||||||||||")
    veces_comer = int(input('Veces que tiene que comer cada filosofo: '))

    filosofos_mesa = 5 #Filosofos sentados en la mesa

    filosofos_comiendo = [] #Lista de filosofos comiendo
    filosofos = [] #Hilos de filosofos
    comiendo = queue.Queue(2) #Cola de filosofos comiendo
    filosofo_comer = [] #Cuantas veces comieron los filosofos

    for i in range(filosofos_mesa): 
        filosofo = Filosofo(i) #Crear hilo de filosofo
        
        filosofos.append(filosofo) #Agregar hilo de filosofo a la lista
        filosofo_comer.append(0) #Agregar cero veces a veces que ha comido el filosofo
        filosofo.start() #Iniciar hilo
    
    for filosofo in filosofos: 
        filosofo.join() 

    print("++++++++++++++++++++++++++++++++++++++++++++++++++")
    print("VECES QUE COMIO FILOSOFO 1: ", filosofo_comer[0])
    print("VECES QUE COMIO FILOSOFO 2: ", filosofo_comer[1])
    print("VECES QUE COMIO FILOSOFO 3: ", filosofo_comer[2])
    print("VECES QUE COMIO FILOSOFO 4: ", filosofo_comer[3])
    print("VECES QUE COMIO FILOSOFO 5: ", filosofo_comer[4])
    print("++++++++++++++++++++++++++++++++++++++++++++++++++")
```
